## 09-03 Ho-Ren-So

Horenso es una crónimo de Hokoku (Informe), Renraku (Comunicación) y Sodan (Consulta), elementos necesarios para que el trabajo se desarrolle sin problemas.

* Hokoku 報告: Es comunicar la información necesaria a superiores o clientes en respuesta a una instrucción, orden o solicitud.
* Renraku 連絡: Es transmitir de forma proactiva la información necesaria a las personas, departamentos u organizaciones pertinentes.
* Sodan 相談: Es solicitar la opinión de personas con capacidad de decisión sobre asuntos que uno mismo no puede resolver.

## 09-05 Pricipio de los tres Gen / cinco Gen

### 1. Principio de los tres Gen

La filosofía de las Tres Realidades (3-Gen Shugi) es la idea de que, al resolver problemas o realizar actividades de mejora, es crucial actual desde las perspectivas del lugar real (Genba), el objeto real (Genbutsu), y la situación real (Genjitsu).


* Genba （現場）: El lugar real (donde ocurre el trabajo o el problema)
* Genbutsu （現物）: El objeto real (el producto, la pieza, etc.)
* Genjitsu （現実）: La cituación real (los hechos y datos concretos).

### 2. Principio de los cinco Gen

La filosofía de las Cinco Realidades (5-Gen Shugi) es la idea de que, para descubrir la causa raíz de un problema y realizar actividades de mejora más efectivas, es crucial adoptar cinco perspectivas: las tres de la filosofía de las Tres Realidades (lugar real, objeto real, situación real), a las que se añaden el principio (Genri) y el precepto (Gensoku).

* Genri 原理: El principio (la teoría o la causa fundamental)
* Gensoku 原則: La regla (la norma o el estándar)

## 09-09 Reglas y Estándares

Las reglas son asuntos que deben cumplirse y que están definidos o documentados.

Los estándares son acuerdos establecidos por la industria o por organizaciones como las empresas. A través de los estándares, las cosas se simplifican y se unifican en diversas situaciones y entre diferentes usuarios.

[Audio_10](https://drive.google.com/file/d/1qA36rjgV2cfKsdrLhBrXWdZJEpRbJ-du/view?usp=drive_link)

```python
from IPython.display import display
from PIL import Image
import gdown
import re

def get_file_id(url,out):
    # 正規表現でファイルIDを抽出
    match = re.search(r'/d/([a-zA-Z0-9_-]+)', url)

    if match:
        file_id = match.group(1)
        print("ファイルID:", file_id)
    else:
        print("ファイルIDが見つかりませんでした。")

    url = f"https://drive.google.com/uc?id={file_id}"

    output = out

    gdown.download(url, output, quiet=False)

    img = Image.open(out)
    display(img)
```
