## 04-02 Proceso anterior y siguiente

Un producto no se fabrica en una sola etapa, sino a través de la conexión de varias etapas. A las etapas directamente conectadas antes y después de cada etapa se les denomina, respectivamente, proceso anterior y proceso siguiente. Dado que el resultado del propio proceso influye en el resultado del proceso siguiente, es importante tener la conciencia de que el proceso siguiente es el cliente y, por lo tanto, asumir la responsabilidad de la calidad en el propio proceso.

### 1. El proceso anterior y el proseso siguiente

Antes de que un producto se termine de fabriar, para por una serie de procesos interconectados.
En esta cadena, el proceso directamente anterior al propio se conoce como **proceso previo** o **proceso anterior**, meintras que e proceso inmediatamente posterior se llma **procesos posterior**.

El producto final (output) del proceso anterior se convierte en el producto de entrada (input) del proceso propio, y el producto final del proceso propio se convierte en el producto de entrada del proceso posterior.
Por lo tanto, si hay un problema en el proceso anterior, será dificil crear un producto de buena calidad en el proceso propio; y si hay un problema en el proceso propio, será difícil crear un producto de buena calidad en el proceso posterior.

Como forma de pensar sobre el proceso, existe la frase **la calidad se construye en el proceso**.
El solo hecho de inspeccionar el resultado del producto en la fase de inspección no se traduce en una mejor calidad.
Es importante tener la conciencia y la acción de que cada persona en cada proceso asume la responsabilidad de entregar un buen producto o servico al proceso siguiente.


### 2. El siguiente proceso es el cliente

Es la idea de procesar correctamente el propio trabajo y entregar los resultados al proceso posterior de la misma manera que uno trataría a un cliente.

## 04-03 Las 5Ms del Proceso

Como elementos importantes para que una línea de producción funcione correctamente y para controlar la calidad del producto, son cruciales cinco factores: Persona (Man), Maquinaria/Equip (Machine), Materia Prima (Material), Método (Method) e Inspección/Medición (Measurement). Tomando la inicial de cada uno, se les conoce como las 5M del proceso.

### (1) Persona (Man)
Si la habilidad del los trabajadores varia, la calidad del producto también variará.

* Hay que proporcionar la capacitación adecuada para las tareas de cada persona.

* Hay que establecer las reglas y hacer entender a las personas su propósito, así como las pérdidas y problemas que surgen si no siguen.

### (2) Maquinaria/Equipo (Machine)
Si el estado de las máquinas y los equipos no es bueno, la calidad del producto puede variar.

* Antes de comenzar el trabajo, es fundamental verificar las condiciones y la precisión de la maquinaria y el equipo.

* Plasmar la configuración de las máquinas y los equipos en los estándares de trabajo.

* Llevar a cabo revisiones y mantenimiento periódico y apropiado de la maquinaria y los equipos.

### (3) Material Prima (Material)

Si hay variaciones o defectos en los materiales y componentes, la calidad del producto final tendrá variaciones o defectos.

* En la fase de recepción de materiales y componentes, hay que verificar que las características estén dentro de las especificaciones.

* En caso de que haya problemas con la calidad de los materiales o componentes, se debe contactar a los proveedores a través del Departamento de Materiales, el Departamento de Compras o el Departamento de Aseguramiento de la Calidad para exigirles medidas de prevención de recurrencia y mejoras en la calidad.

### (4) Método (Method)

Si los procedimientos de trabajo para el procesamiento y el ensamblaje varían en sus condiciones y métodos, la calidad del producto tendrá variaciones.

* Establecer los procedimientos y los métodos de trabajo y crear estándares de trabajo y diagramas de flujo de control de calidad.

* Educar y capacitar para que el trabajo se realice de acuerdo con los estándares de trabajo.

* Cuando el método de trabajo se modifique para mejorar la calidad, es fundamental revisar el estándar de trabajo y el diagrama de flujo de control de calidad.

### (5) Inspección/Medición (Measurement)

Las variaciones en la precisión de  los instrumentos de medición, las condiciones y métodos de medición, o las habilidades de los operadores, pueden causar variaciones en los datos de medición.

* Llevar a cabo revisiones y calibraciones de los instrumentos de medición para reducir el sesgo y la variabilidad.

* Estandarizar los métodos de medición de los instrumentos de medición y los métodos para identificar si un producto es aceptable o no.

* Brindar educación y capacitación para mejorar las habilidades delos operadores de medición.

Las **5M del proceso** son elementos cruciales que no se deben pasar por alto en la gestión de la calidad durante la fabricación de un producto.
Las 5M son los elementos básicos para llevar a cabo la gestión de la calidad de forma adecuada en el entorno de la producción y los servicios.

## 04-04 Anomalía

Se refiere a la condición en la que un proceso o un producto, debido a alguna causa, se encuentra en un estado diferente al normal.
Las causas de la variabilidad en las características de un producto se dividen en **causas asignables** (o especiales) y **causas comunes** (o aleatorias), pero una **anomalía** se refiere a aquella causada por una **causa assignable**.

Aunque un proceso esté correctamente gestionado, siempre tendrá variabilidad.
Existen diversas causas de la variabilidad.

**Causas comunes**:
las causas inevitables que ocurren incluso cuando el trabajo se realiza correctamente se denominan causas comunes (causas de variación comunes).

**Causas assignable**:
Las causas debidas a alguna anomalía, como defectos en el trabajo o fallos en el equipo, se denominan **causas asignable (causas especiales)**.

Las causas especiales pueden hacer que la **media de** un proceso cambie o que su **variabilidad** se modifique.

Las causas asignable son causas que no
se pueden pasar por alto y que deben ser abordadas y eliminadas de forma definitiva.

```python
from IPython.display import display
from PIL import Image
import gdown
import re

def get_file_id(url,out):
    # 正規表現でファイルIDを抽出
    match = re.search(r'/d/([a-zA-Z0-9_-]+)', url)

    if match:
        file_id = match.group(1)
        print("ファイルID:", file_id)
    else:
        print("ファイルIDが見つかりませんでした。")

    url = f"https://drive.google.com/uc?id={file_id}"

    output = out

    gdown.download(url, output, quiet=False)

    img = Image.open(out)
    display(img)
```
