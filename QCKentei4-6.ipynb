## 06-01 Estandarización / Normalización （標準化）

Se refiere a la actividad de establecer como estándar aquello que se considera el mejor método para realizar algo (ya sea un objeto o una forma de trabajar), y actuar de acuerdo con dicho estándar.

Es necesario establecer reglas en las actividades empresariales para que varias personas puedan actuar de forma precisa, eficiente, sin desperdicios y sin problemas.
El proceso de establecer estas reglas y actuar de acuerdo con ellas se denomina **estandarización**（標準化）.

Un **estándar (o norma)**（標準）es un acuerdo establecido con el propósito de lograr la unificación y la amplificación, de manera que los beneficios o la conveniencia se obtengan de forma equitativa entre las personas u organizaciones relacionadas.

Los estándares documentados se conoce generalmente como **documentos de estándar** o **procedimientos estándar**（標準書）, y describen la forma más apropiada de realizar un trabajo o una tarea.
Si el trabajo o la tarea se llevan a cabo siguiendo estos documentos, se pueden prevenir los errores y avanzar de manera eficiente.

## 06-02 Criterios relacionados con el trabajo
Estándares relacionados con las operaciones / Normas relativas a las tareas. Estándares para articulos (especificaciones ) / Normas para productos (especificaciones) / Criterios para bienes (especificaciones)

### 1. Criterios relacionados con el trabajo
Estándares relacionados con las operaciones / Normas relativas a las tareas.Estándares para articulos (especificaciones ) / Normas para productos (especificaciones) / Criterios para bienes (especificaciones)

En general, una **especificación** es un tipo de estándar que establece acuerdos técnicos relacionados directa o indirectamente con un producto, servicio, proceso o sistema.

Los acuerdos que establecen principalmente los detalles, procedimientos, trámites y métodos de una organización o de sus operaciones se denominan **normas** o **reglamenteos**.

A las normas establecidas dentro de una empresa sobre su funcionamiento y sus productos se les llama **estándares internos**.
Estos estándares son reglas que deben seguirse para que las actividades de la empresa se lleven a cabo de manera apropiada y racional.

### 2. Estándar interno （社内標準）

1. Regulación, Disposición, Norma, Reglamento （規定）
  * Es el conjunto de reglas establecidas sobre el contenido, los procedimientos, los trámites y los métodos de las diferentes operaciones que se llevan a cabo dentro de una organización empresarial.

2. Especificación, Estándar, Norma, Pliego de condiciones.（規格）
  * Esto se refiere a un documento que establece los asuntos técnicos relacionados con productos, productos semiacabados, piezas, materiales, entre otros.
  
3. Estándar （標準）
  * Documento que establece los criterios para los métodos de trabajo en el lugar de producción y en la oficina, así como los métodos de gestión y las precauciones (estándares de trabajo), o que establece los asuntos técnicos necesarios par alas tareas de diseño y desarrollo (estándares técnicos).

4. Procedimiento, Manual （手順書・マニュアル）
  * Un documento que detalla los estándares de trabajo y los estándares técnicos para cada tarea, desgloseándolos en procedimientos y manuales para el personal a cargo.

5. Especificaciónes técnicas, Pliego de especificaciones, Ficha técnica. （仕様書）
  * Documento que especifica la forma, la estructura, las dimensiones, la composición, la capacidad, la precisión, el rendimiento, el método de fabricación, el método de prueba, etc., requeridos para materiales, productos, herramientas y equipos.

```python
from IPython.display import display
from PIL import Image
import gdown
import re

def get_file_id(url,out):
    # 正規表現でファイルIDを抽出
    match = re.search(r'/d/([a-zA-Z0-9_-]+)', url)

    if match:
        file_id = match.group(1)
        print("ファイルID:", file_id)
    else:
        print("ファイルIDが見つかりませんでした。")

    url = f"https://drive.google.com/uc?id={file_id}"

    output = out

    gdown.download(url, output, quiet=False)

    img = Image.open(out)
    display(img)
```
