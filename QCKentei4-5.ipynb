## 05-02 Conforme (Producto)

**Conforme (Producto)**: Es el estado (o artículo) que, tras la inspección, satisface los requisitos especificados en las exigencias del cliente o en las normas.

**No conforme (Producto)**: Es el estado (o artículo) que, tras la inspección, no satisface los requisitos especificados en las exigencias del cliente o en las normas.

## 05-03 Aprobación/Rechazo de Lotes

Lotes: Se define como lote a una agrupación de articulos que han sido fabricados, o se presume que lo han sido, bajo las mismas condiciones.


### 1. Lotes

Por ejemplo, un grupo de 100 piezas de precisión producidas en la misma máquina en un solo día puede considerarse un solo lote, ya que comparten las mismas características: fueron fabricadas el mismo día, en la misma máquina y por el mismo operario.

El número de productos dentro de un solo lote se denomina **tamaño del lote**.

Se puede pensar en los lotes también en el contexto de los servicios.
Por ejemplo, en un centro de llamadas, el conjunto de consultas atendidas por un operador en un día determinado puede considerarse un lote.

### 2. Inspección de lotes / Control de lotes / Verificación de lotes

Es frecuente que la inspección de lotes se realice extrayendo una **muestra** del lote, realizando mediciones y pruebas sobre dicha muestra, y luego determinando si cumple con los criterios preestablecidos.

A este tipo de inspección se le llama **inspección por muestreo**.

Además, la decisión se toma para el lote completo: si cumple con los criterios, se considera aprobado (o aceptado); si no, se considera rechazado.

Hay que tener en cuenta que, aunque las mediciones y pruebas se realizan sobre la muestra, la decisión de aprobación o rechazo se aplica al lote completo.

### 2. Inspección por método / Control según el método / Verificación por método

Entre los métodos de inspección se encuentran la inspección 100%, la inspección por muestreo y la inspección sin pruebas.

#### (1) Inspección 100%

* Cuando no se premite incluir ni siquiera un solo producto no
conforme / Cuando la inclusión de un solo producto no conformes es inaceptable
* Cuando está en en juesgo la vida humana / En casos que involucran vidas humanas.
* Cuando las pérdidas económicas son muy grandes / En caso de pérdidas económicas significativas.

#### (2) Inspección muestreo

La inspección en la que se determina si un lote completo es aceptado o rechazado basándose en una muestra tomada de él se denomina inspección por muestreo.

Dado que la inspección por muestreo no examina todos los productos de un lote, es posible que un lote aceptado aún contenga productos no conformes.

Por lo tanto, este método se aplica cuando el cliente acepta que pueda haber una cierta cantidad de productos no conformes en el envío.

#### (3) Inspección sin pruebas

La inspección que determina la conformidad basándose únicamente en la documentación presentada, sin realizar mediciones o pruebas, se denomina inspección sin pruebas.


### 3. Inspección por tipo de propiedad / Inspección según la naturaleza / Inspección por característica

Entre los tipos de inspección según la naturaleza, se encuentran la inspección destructiva, la inspección no destructiva y la inspección sensorial

#### (1) Inspección destructiva

La inspección en la que el objeto de prueba es destruido, total o parcialmente, perdiendo su funcionalidad, se denomina inspección destructiva.

#### (2) Inspección no destructiva

La inspección que se realiza sin destruir el objeto durante las mediciones o pruebas se denomina inspección no destructiva.

#### (3) Inspección sensorial

La inspección que utiliza los cinco sentidos humanos (vista, oido, gusto, olfato y tacto) durante las mediciones o pruebas se denomina inspección sensorial o inspección organoléptica.

### 4. Inspección alternativa / Inspección de reemplazo

La inspección sustitutiva (o alternativa) es otra característica de calidad que se utiliza como una evaluación equivalente o aproximada cuando resulta difícil medir directamente la característica de calidad requerida.

```python
from IPython.display import display
from PIL import Image
import gdown
import re

def get_file_id(url,out):
    # 正規表現でファイルIDを抽出
    match = re.search(r'/d/([a-zA-Z0-9_-]+)', url)

    if match:
        file_id = match.group(1)
        print("ファイルID:", file_id)
    else:
        print("ファイルIDが見つかりませんでした。")

    url = f"https://drive.google.com/uc?id={file_id}"

    output = out

    gdown.download(url, output, quiet=False)

    img = Image.open(out)
    display(img)
```
